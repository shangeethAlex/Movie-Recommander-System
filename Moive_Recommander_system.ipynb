```python
### merge data 
movies = movies.merge(credits,on='title')
movies.shape
```

```python
## select specific colums to use into recommander
movies = movies[['movie_id','title','cast','overview','genres','keywords','crew']]
movies.head()
```

### create covert function 

- Uses ast.literal_eval to safely convert string representations of lists of dictionaries into actual Python lists.

```python
import ast

def convert(text):
    L= []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

### Drop missing values

```python
movies.dropna()
```

### apply convert function to genre column

-  Applies the convert function to each element in the genres column of the movies DataFrame.

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies.cast[1]
```

### limit the cast information to the first three cast members for each movie in the cast column. 

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
movies.head()
```

## check the crew column

```python
movies.crew[1]
```

- The fetch_director function extracts the names of directors from the crew column,

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == "Director":
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head()
```

#### removes spaces from each string in the lists within the specified columns of the movies DataFrame.

```python
def colapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
#cast #crew #genres #keywords
```

```python
movies['cast'] = movies['cast'].apply(colapse)
movies['crew'] = movies['crew'].apply(colapse)
movies['genres'] = movies['genres'].apply(colapse)
movies['keywords'] = movies['keywords'].apply(colapse)
```

```python
movies.head()
```

#### transforms the overview column of the movies DataFrame so that each overview is represented as a list of words.

```python
movies['overview'] = movies['overview'].fillna('')
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(2)
```

```python
movies['tags'] = movies['overview'] + movies['keywords'] + movies['crew'] + movies['cast'] + movies['genres']
movies.head()
```

```python
new_df = movies.drop(columns=['overview','keywords','crew','cast','genres'])
new_df.head(2)
```

#### convert each list of tags in the tags column of the DataFrame new into a single string where tags are separated by spaces.

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
new_df.head(2)
```

### recommander system

 - The cosine_similarity function is imported from sklearn.metrics.pairwise to compute the cosine similarity between the vectors.
 - countVectorizer is imported from sklearn.feature_extraction.text to convert a collection of text documents to a matrix of token counts

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
vector.shape
```

- The cosine similarity matrix is computed for the vectors. This matrix indicates the similarity between each pair of movies based on their tags.

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommanded(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    print(f"recomandation for {movie} :")
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
    
```

```python
recommanded('Gandhi')
```
